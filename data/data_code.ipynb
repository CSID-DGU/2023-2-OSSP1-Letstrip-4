```python
import pandas as pd
metadata = pd.read_csv('C:/Users/user/OneDrive/바탕 화면/movie_metadata.csv')
metadata.head(2)
```

```python
image_data = metadata[['movie_imdb_link','movie_title','genres','summary','director_name','actor_1_name','actor_2_name','actor_3_name']]
image_data.head(10)
```

```python
links = pd.read_csv('C:/Users/user/OneDrive/바탕 화면/links.csv')
links.head(10)
```

```python
links = links[['movieId','imdbId']]
links.head(10)
```

```python
image_data = image_data[ ~ image_data.movie_imdb_link.isnull()]
def app(x):
 try:
    return int(x[28:35])
 except ValueError:
    print (x)
image_data['imdbId'] = image_data.movie_imdb_link.apply(app)
image_data = image_data[ ~ image_data.imdbId.isnull()]
image_data.imdbId = image_data.imdbId.astype(int)
image_data = image_data[['imdbId','movie_title','genres','summary','director_name','actor_1_name','actor_2_name','actor_3_name']]
image_data.head(10)
```

```python
posters = pd.merge(image_data, links, on='imdbId', how='left')
posters = posters[['movieId','movie_title','genres','summary','director_name','actor_1_name','actor_2_name','actor_3_name']]
posters = posters[~ posters.movieId.isnull()]
posters.movieId = posters.movieId.astype(int)
posters.head(10)
```

```python
title = pd.read_csv('C:/Users/user/OneDrive/바탕 화면/movies.csv')
title.head(10)
```

```python
title = title[['movieId','title']]
title.head(10)
```

```python
A = pd.merge(posters, title, on='movieId', how='left')
A = A[['movieId','title','genres','director_name','actor_1_name','actor_2_name','actor_3_name']]
A = A[~ A.title.isnull()]
A.title = A.title.astype(str)
A.head(10)
```

```python
overview = pd.read_csv('C:/Users/user/OneDrive/바탕 화면/overview.csv')
overview.head(2)
```

```python
overview = overview[['movieId','overview']]
overview.head(10)
```

```python
final = pd.merge(A, overview, on='movieId', how='left')
final.head(10)
```

```python
final.to_csv("final.csv")
```
